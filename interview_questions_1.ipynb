```python
# https://leetcode.com/problems/max-points-on-a-line/

# Given an array of points where points[i] = [xi, yi] represents a point on the X-Y plane, return the maximum number of points that lie on the same straight line.

class Solution(object):
    def maxPoints(self, points):
        if len(points) <= 2:
            return len(points)
        
        points = sorted(points, key=lambda k:(k[0], k[1]))
        f = [{} for i in range(len(points))]
        max_len = 0
        
        for i in range(1, len(points)):
            for j in range(i):
                s = float((points[i][1]-points[j][1]))/(points[i][0]-points[j][0]) if points[i][0]-points[j][0] != 0 else float("Inf")
                
                if s in f[j]:
                    if s not in f[i]:
                        f[i][s] = 0
                    f[i][s] = max(f[i][s], f[j][s]+1)
                else:
                    f[i][s] = 2
                
                max_len = max(max_len, f[i][s])
        
        return max_len
```

```python
# https://leetcode.com/discuss/interview-question/233608/Uber-or-Maximum-drivers-online

# Give you an extra long list. Each element of the list is an array. For example: [[1, 5], [10, 16], [2, 7],..... [1000, 1027]]

# The first number of arrays represents the driver's starting time and the second number is the arrival time for passengers. Question: Given the starting and ending time for online drivers from this input array, find the time interval where there is maximum drivers online. For example, the answer could be [16, 801]

def max_drivers_online(intervals):
    intervals = sorted(intervals)
    heap = []
    max_online, start, end = 0, -1, -1
    
    for i in range(len(intervals)):
        if len(heap) > 0 and heap[0] < intervals[i][0]:
            heapq.heappop(heap)
        heapq.heappush(heap, intervals[i][1])
        
        if len(heap) > max_online:
            max_online = len(heap)
            start, end = intervals[i][0], heap[0]
    
    return max_online, start, end
```

```python
# https://leetcode.com/problems/remove-invalid-parentheses/

# Given a string s that contains parentheses and letters, remove the minimum number of invalid parentheses to make the input string valid.

# Return all the possible results. You may return the answer in any order.

class Solution(object):
    def remove_invalid(self, s, i, left, right, curr, visited, out):
        if i == len(s):
            if left == right:
                if left not in out:
                    out[left] = set()
                out[left].add(curr)
                
        else:
            if s[i] == '(':
                if left >= right:
                    if curr + '(' not in visited:
                        visited.add(curr + '(')
                        self.remove_invalid(s, i+1, left+1, right, curr + '(', visited, out)
                    
                    self.remove_invalid(s, i+1, left, right, curr, visited, out)
                    
                elif left == right-1:
                    if curr + '(' not in visited:
                        visited.add(curr + '(')
                        self.remove_invalid(s, i+1, left+1, right, curr + '(', visited, out)
                
            elif s[i] == ')':
                if right < left:
                    if curr + ')' not in visited:
                        visited.add(curr + ')')
                        self.remove_invalid(s, i+1, left, right+1, curr + ')', visited, out)
                        
                    self.remove_invalid(s, i+1, left, right, curr, visited, out)
                    
                elif right == left:
                    self.remove_invalid(s, i+1, left, right, curr, visited, out)
                    
            else:
                if curr + s[i] not in visited:
                    visited.add(curr + s[i])
                    self.remove_invalid(s, i+1, left, right, curr + s[i], visited, out)
        
        
    def removeInvalidParentheses(self, s):
        out = {}
        self.remove_invalid(s, 0, 0, 0, '', set(['']), out)
        max_len = max([k for k in out])
        return out[max_len]
```

```python
# https://leetcode.com/discuss/interview-question/323701/Uber-or-Onsite-or-Split-array-into-k-subarrays-with-min-abs-diff

# Given an array arr of non-negative integers. You need to split it into k contiguous subarrays such that the absolute difference between max sum and min sum is the smallest and output that difference.

# Example:

# Input: arr = [3, 4, 5, 3, 7, 2], k = 3
# Output: 2
# Explanation: Optimal split is subarray1 = [3, 4], subarray2 = [5, 3], subarray3 = [7, 2]
# Min sum = 7 and max sum = 9, hence the answer is |7 - 9| = 2.

def min_abs_difference(arr, k):
    left, right = min(arr), sum(arr)
    p = -1
    res = 0
    
    while left <= right:
        s = int((left + right)/2)
        
        i = 0
        h = 0
        parts = 0
        min_sum, max_sum = float("Inf"), -float("Inf")
        
        while i < len(arr):
            if h+arr[i] > s:
                parts += 1
                
                min_sum = min(min_sum, h)
                max_sum = max(max_sum, h)
                
                h = arr[i]
            else:
                h += arr[i]
            
            i += 1
        
        parts += 1
        
        min_sum = min(min_sum, h)
        max_sum = max(max_sum, h)
        
        if parts >= k:
            p = s
            left = s+1
            res = abs(max_sum-min_sum)
        else:
            right = s-1
    
    return res

min_abs_difference([3, 4, 5, 3, 7, 2], 3)
```

```python
# https://leetcode.com/discuss/interview-question/332346/Uber-or-Onsite-or-Crop-Farming

# Given cost and profit array of farming pattern. Maximize profit for m months. For capital x only crops with cost below or equal to x can be farmed. Each month you can farm only one crop. Once a crop is farmed it can't be used again. After each crop the total capital becomes x + profit.

# Example 1:

# Input: cost = [1, 8, 9], profit = [2, 3, 8], capital = 7, m = 2
# Output: 17
# Explanation: In the begining only crop with cost 1 can be farmed. After first month capital = 7 + 2 = 9.
# Next crop 9 should be farmed as profit is max for that to give 9 + 8 = 17. So max profit after 2 months will be 17.

import heapq

def crop_farming(cost, profit, capital, m):
    heap = [(cost[i], -profit[i], i) for i in range(len(cost))]
    heapq.heapify(heap)
    
    while m > 0:
        max_profit = 0
        
        while len(heap) > 0 and capital >= heap[0][0]:
            c, p, i = heapq.heappop(heap)
            max_profit = max(max_profit, -p)
        
        capital += max_profit
        m -= 1
    
    return capital
```

```python
# https://www.lintcode.com/problem/307/

# Given two strings A and B, the task is to convert A to B if possible. The only operation allowed is to put any character from A and insert it at front. Find if it’s possible to convert the string. If yes, then output minimum number of operations required for transformation.If no, then output -1.
# Hint:0 ≤ the length of A, the length of B ≤ 100000

# Example
# Example 1:

# Input: A = "ABD", B = "BAD"
# Output: 1
# Explanation：Pick B and insert it at front.

class Solution:
    """
    @param A: a string A
    @param B: a string B
    @return: return the minimum number of operations to transform
    """
    def transform(self, A, B):
        if len(A) != len(B):
            return -1
        
        a, b = [0]*57, [0]*57
        
        for i in range(len(A)):
            a[ord(A[i])-ord('A')] += 1
            b[ord(B[i])-ord('A')] += 1
        
        if tuple(a) != tuple(b):
            return -1
            
        i, j = len(A)-1, len(B)-1
        cnts = 0
        
        while i >= 0 and j >= 0:
            if A[i] == B[j]:
                i -= 1
                j -= 1
                cnts += 1
            else:
                i -= 1
        
        return len(A)-cnts
```

```python
# https://www.lintcode.com/problem/1532/

"""
Definition for a directed graph node
class DirectedGraphNode:
    def __init__(self, x):
        self.label = x
        self.neighbors = []
"""


class Solution:
    def serialize(self, root):
        if root is None:
            return []

        out = [root.label]

        for x in root.neighbors:
            a = self.serialize(x)
            out += [len(a)] + a
        
        return out
        
    def deserialize(self, data):
        root = DirectedGraphNode(data[0])

        j = 1
        while j < len(data):
            n = data[j]
            new_data = data[j+1:j+n+1]
            x = self.deserialize(new_data)
            root.neighbors += [x]
            j = j+n+1
        
        return root
```

```python
# https://leetcode.com/discuss/interview-question/297202/Uber-or-Onsite-or-Surround-substrings

# You have two arrays of strings, words and parts. Write a function findSubstrings(words, parts) that returns the array that contains the strings from words, modified so that any occurrences of the substrings from parts are surrounded by square brackets [], following these guidelines:

# If several parts substrings occur in one string in words, choose the longest one.
# If there is still more than one such part, then choose the one that appears first in the string.
# Example 1:

# Input: words = ["Apple", "Melon", "Orange", "Watermelon"], parts = ["a", "mel", "lon", "el", "An"]
# Output: ["Apple", "Me[lon]", "Or[a]nge", "Water[mel]on"]. 
# Explanation: While "Watermelon" contains three substrings from the parts array ("a", "mel", and "lon") "mel" is the longest substring
# that appears first in the string.

class Trie:
    def __init__(self):
        self.char_map = {}
    
    def insert(self, word):
        if len(word) == 0:
            self.char_map['$'] = Trie()
        else:
            if word[0] not in self.char_map:
                self.char_map[word[0]] = Trie()
            
            self.char_map[word[0]].insert(word[1:])
    
    def search(self, s, i):
        y = -1
        
        if '$' in self.char_map:
            y = i-1
            
        if i < len(s) and s[i] in self.char_map:
            return max(y, self.char_map[s[i]].search(s, i+1))
        
        return y

def surround(words, parts):
    trie = Trie()
    for part in parts:
        trie.insert(part)
    
    out = []
    
    for word in words:
        m = 0
        p, q = 0, -1
        
        for i in range(len(word)):
            j = trie.search(word, i)
            if j-i+1 > m:
                m = j-i+1
                p, q = i, j
                
        if p <= q:
            out.append(word[:p] + '(' + word[p:q+1] + ')' + word[q+1:])
        else:
            out.append(word)
    
    return out
```

```python
# https://leetcode.com/discuss/interview-question/374697/Uber-or-Phone-Screen-or-Tournament

# Given an array arr of unique integers [1..n]. The array represents n different players at a tournament and their overall strength, and the first two slots in the array represent the current match. The stronger player always wins and the weaker player is moved to the end of the array. The tournament ends when a player wins k matches in a row. Find the winner.

# Example:

# Input: arr = [2, 1, 3, 4, 5], k = 2
# Output: 5
# Explanation:
# [2 1] 3 4 5 | player 2 vs. player 1, player 2 wins and player 1 is moved to the end
# [2 3] 4 5 1 | player 2 vs. player 3, player 3 wins
# [3 4] 5 1 2 | player 4 wins
# [4 5] 1 2 3 | player 5 wins
# [5 1] 2 3 4 | player 5 wins again, that's k times in a row so 5 is the overall winner

def tournament(arr, k):
    i = 0
    
    while i < len(arr):
        j = i+1
        h = 0 if i == 0 else 1
        
        while j < len(arr) and arr[j] < arr[i]:
            h += 1
            j += 1
        
        if h >= k:
            return arr[i]
        
        i = j
        
    return arr[-1]
```

```python
# https://www.lintcode.com/problem/918/

# Given an array of n integers nums and a target, find the number of index triplets i, j, k with 0 <= i < j < k < n that satisfy the condition nums[i] + nums[j] + nums[k] < target.

# Example
# Example1

# Input:  nums = [-2,0,1,3], target = 2
# Output: 2
# Explanation:
# Because there are two triplets which sums are less than 2:
# [-2, 0, 1]
# [-2, 0, 3]

class Solution:
    """
    @param nums:  an array of n integers
    @param target: a target
    @return: the number of index triplets satisfy the condition nums[i] + nums[j] + nums[k] < target
    """
    def threeSumSmaller(self, nums, target):
        if len(nums) < 3:
            return 0

        nums = sorted(nums)
        cnts = 0

        cache = [0]*len(nums)

        for i in range(len(nums)):
            cache[i] = target-nums[i]

        for i in range(1, len(nums)):
            k = len(nums)-1
            for j in range(i):
                s = nums[i] + nums[j]
                while k > i and cache[k] <= s:
                    k -= 1
                cnts += k-i
        
        return cnts
```

```python
# https://leetcode.com/discuss/interview-question/360783/Uber-ATG-or-Onsite-or-N-ary-Tree-Reconstruction

# Given an unordered list of pairs where each pair represents a parent-child relationship (not nessesary direct) in an N-ary tree. Reconstruct and return the original tree. You can assume that each node has a unique value.

# Example 1:

# Input: [[1, 2], [1, 3], [1, 4], [2, 4]]
# Output:
#      1
#     / \
#    2   3
#   /
#  4

class TreeNode:
    def __init__(self, val):
        self.val = val
        self.neighbors = []
        
def construct_n_ary_tree(n, edges):
    adj_map = [[] for i in range(n)]
    in_deg = [0]*n
    
    for x, y in edges:
        adj_map[x].append(y)
        in_deg[y] += 1
    
    arr = []
    root = None
    
    for i in range(n):
        if in_deg[i] == 0:
            root = TreeNode(i)
            arr.append((root, i))
            break
    
    while len(arr) > 0:
        new_arr = []
        
        for node, i in arr:
            if i in adj_map:
                for j in adj_map[i]:
                    in_deg[j] -= 1
                    if in_deg[j] == 0:
                        new_node = TreeNode(j)
                        node.neighbors += [new_node]
                        new_arr.append((new_node, j))
        
        arr = new_arr[:]
    
    return root
```

```python
# https://leetcode.com/problems/number-of-ways-to-reorder-array-to-get-same-bst/

# Given an array nums that represents a permutation of integers from 1 to n. We are going to construct a binary search tree (BST) by inserting the elements of nums in order into an initially empty BST. Find the number of different ways to reorder nums so that the constructed BST is identical to that formed from the original array nums.

# For example, given nums = [2,1,3], we will have 2 as the root, 1 as a left child, and 3 as a right child. The array [2,3,1] also yields the same BST but [3,2,1] yields a different BST.

# Return the number of ways to reorder nums such that the BST formed is identical to the original BST formed from nums.

class Solution(object):
    def fn(self, n, m, cache):
        if n == 0 or m == 0:
            return 1

        a = self.fn(n-1, m, cache) if (n-1, m) not in cache else cache[(n-1, m)]
        b = self.fn(n, m-1, cache) if (n, m-1) not in cache else cache[(n, m-1)]

        cache[(n, m)] = a + b
        return a + b

    def num_permutation_bsts(self, arr, cache):
        l = [arr[i] for i in range(1, len(arr)) if arr[i] < arr[0]]
        r = [arr[i] for i in range(1, len(arr)) if arr[i] > arr[0]]

        a = self.num_permutation_bsts(l, cache) if len(l) > 0 else 1
        b = self.num_permutation_bsts(r, cache) if len(r) > 0 else 1

        return a*b*self.fn(len(l), len(r), cache)

    def numOfWays(self, nums):
        return (self.num_permutation_bsts(nums, {})-1) % (10**9+7)
```

```python
# https://www.lintcode.com/problem/599/

# Given a node from a cyclic linked list which has been sorted, write a function to insert a value into the list such that it remains a cyclic sorted list. The given node can be any single node in the list. Return the inserted new node.

# 3->5->1 is a cyclic list, so 3 is next node of 1.
# 3->5->1 is same with 5->1->3
# Example
# Example 1:

# Input:
# 3->5->1
# 4
# Output:
# 5->1->3->4

class Solution:
    """
    @param node: a list node in the list
    @param x: An integer
    @return: the inserted new list node
    """
    def insert(self, node, x):
        if node is None:
            g = ListNode(x)
            return g
            
        ptr = node

        while True:
            if ptr.val <= x <= ptr.next.val or (ptr.next.val < ptr.val and x < ptr.next.val and x < ptr.val) or ptr.next == node:
                g = ListNode(x)
                ptr.next = g
                g.next = ptr.next
                break

            ptr = ptr.next
        
        return node
```

```python
# https://leetcode.com/problems/decode-string/

# Given an encoded string, return its decoded string.

# The encoding rule is: k[encoded_string], where the encoded_string inside the square brackets is being repeated exactly k times. Note that k is guaranteed to be a positive integer.

# You may assume that the input string is always valid; No extra white spaces, square brackets are well-formed, etc.

# Furthermore, you may assume that the original data does not contain any digits and that digits are only for those repeat numbers, k. For example, there won't be input like 3a or 2[4].

 

# Example 1:

# Input: s = "3[a]2[bc]"
# Output: "aaabcbc"

class Solution(object):
    def decodeString(self, s):
        stack = []
        p = 0
        q = ''
        
        for i in range(len(s)):
            if s[i].isdigit():
                p = p*10 + int(s[i])
                if len(q) > 0:
                    stack.append(q)
                
                q = ''
                    
            elif s[i] == '[':
                if p > 0:
                    stack.append(str(p))
                if len(q) > 0:
                    stack.append(q)
                    
                p = 0
                q = ''
                
            elif 97 <= ord(s[i]) <= 122:
                q += s[i]
                if p > 0:
                    stack.append(str(p))
                
                p = 0
                    #
            else:
                g = q
                while True:
                    h = stack.pop()
                    if h.isdigit():
                        break
                    g = h + g
                
                x = g*int(h) if h.isdigit() else g
                stack.append(x)
                
                p = 0
                q = ''
        
        if p > 0:
            stack.append(str(p))
            
        if len(q) > 0:
            stack.append(q)
            
        return ''.join(stack)
```

```python
# https://leetcode.com/problems/task-scheduler/

# Given a characters array tasks, representing the tasks a CPU needs to do, where each letter represents a different task. Tasks could be done in any order. Each task is done in one unit of time. For each unit of time, the CPU could complete either one task or just be idle.

# However, there is a non-negative integer n that represents the cooldown period between two same tasks (the same letter in the array), that is that there must be at least n units of time between any two same tasks.

# Return the least number of units of times that the CPU will take to finish all the given tasks.

 

# Example 1:

# Input: tasks = ["A","A","A","B","B","B"], n = 2
# Output: 8
# Explanation: 
# A -> B -> idle -> A -> B -> idle -> A -> B
# There is at least 2 units of time between any two same tasks.

class Solution(object):
    def leastInterval(self, tasks, n):
        cnts = {}
        for x in tasks:
            if x not in cnts:
                cnts[x] = 0
            cnts[x] += 1
        
        heap = [(-cnts[x], x) for x in cnts]
        heapq.heapify(heap)
        last_pos = {}
        occupied = set()
        
        i = 0
        m = 0
        
        while len(heap) > 0:
            while i in occupied:
                i += 1
                
            s, x = heapq.heappop(heap)
            s = -s
            
            if x not in last_pos or i-last_pos[x] > n:
                last_pos[x] = i
                i += 1
            else:
                last_pos[x] += n+1
                
            occupied.add(last_pos[x])
            m = max(m, last_pos[x])
            
            if s-1 > 0:
                heapq.heappush(heap, (-(s-1), x))
        
        return m+1
```

```python
# https://leetcode.com/problems/number-of-submatrices-that-sum-to-target/

# Given a matrix and a target, return the number of non-empty submatrices that sum to target.

# A submatrix x1, y1, x2, y2 is the set of all cells matrix[x][y] with x1 <= x <= x2 and y1 <= y <= y2.

# Two submatrices (x1, y1, x2, y2) and (x1', y1', x2', y2') are different if they have some coordinate that is different: for example, if x1 != x1'.

class Solution(object):
    def numSubmatrixSumTarget(self, matrix, target):
        n, m = len(matrix), len(matrix[0])
        
        sums = [[0]*m for i in range(n)]
        
        for i in range(n):
            s = 0
            for j in range(m):
                s += matrix[i][j]
                sums[i][j] = sums[i-1][j] + s if i > 0 else s
        
        out = 0
        
        for i in range(n):
            for k in range(i, n):
                sum_cache = {0:1}
                s = 0
                for j in range(m):
                    s = sums[k][j]-sums[i-1][j] if i > 0 else sums[k][j]
                    
                    if s-target in sum_cache:
                        out += sum_cache[s-target]
                    
                    if s not in sum_cache:
                        sum_cache[s] = 0
                    sum_cache[s] += 1
                    
        return out
```

```python
# https://leetcode.com/problems/pancake-sorting/

# Given an array of integers arr, sort the array by performing a series of pancake flips.

# In one pancake flip we do the following steps:

# Choose an integer k where 1 <= k <= arr.length.
# Reverse the sub-array arr[0...k-1] (0-indexed).
# For example, if arr = [3,2,1,4] and we performed a pancake flip choosing k = 3, we reverse the sub-array [3,2,1], so arr = [1,2,3,4] after the pancake flip at k = 3.

# Return an array of the k-values corresponding to a sequence of pancake flips that sort arr. Any valid answer that sorts the array within 10 * arr.length flips will be judged as correct.

class Solution(object):
    def pancakeSort(self, A):
        out = []
        
        curr = len(A)
        while curr > 1:
            i = A.index(curr)
            if i < curr-1:
                if i > 0:
                    A[:i+1] = A[:i+1][::-1]
                    out.append(i+1)
                
                A[:curr] = A[:curr][::-1]
                out.append(curr)
                
            curr -= 1
            
        return out
```

```python
# https://leetcode.com/problems/course-schedule-ii/

# There are a total of numCourses courses you have to take, labeled from 0 to numCourses - 1. You are given an array prerequisites where prerequisites[i] = [ai, bi] indicates that you must take course bi first if you want to take course ai.

# For example, the pair [0, 1], indicates that to take course 0 you have to first take course 1.
# Return the ordering of courses you should take to finish all courses. If there are many valid answers, return any of them. If it is impossible to finish all courses, return an empty array.

 

# Example 1:

# Input: numCourses = 2, prerequisites = [[1,0]]
# Output: [0,1]
# Explanation: There are a total of 2 courses to take. To take course 1 you should have finished course 0. So the correct course order is [0,1].

class Solution(object):
    def findOrder(self, numCourses, prerequisites):
        if numCourses == 0:
            return []
        
        adj_list = [[] for i in range(numCourses)]
        in_deg = [0]*numCourses
        
        for x, y in prerequisites:
            adj_list[y].append(x)
            in_deg[x] += 1
        
        arr = []
        for i in range(numCourses):
            if in_deg[i] == 0:
                arr.append(i)
        
        out = []
        
        while len(arr) > 0:
            new_arr = []
            
            for i in arr:
                out.append(i)
                
                for j in adj_list[i]:
                    in_deg[j] -= 1
                    if in_deg[j] == 0:
                        new_arr.append(j)
            
            arr = new_arr[:]
        
        if max(in_deg) > 0:
            return []
        
        return out
```

```python
# https://leetcode.com/problems/kth-missing-positive-number/

# Given an array arr of positive integers sorted in a strictly increasing order, and an integer k.

# Find the kth positive integer that is missing from this array.

 

# Example 1:

# Input: arr = [2,3,4,7,11], k = 5
# Output: 9
# Explanation: The missing positive integers are [1,5,6,8,9,10,12,13,...]. The 5th missing positive integer is 9.

class Solution(object):
    def findKthPositive(self, arr, k):
        left, right = 0, len(arr)-1
        p, h = -1, k
        
        while left <= right:
            mid = int((left + right)/2)
            
            if arr[mid]-(mid+1) < k:
                p = mid
                h = k-(arr[mid]-(mid+1))
                left = mid+1
            else:
                right = mid-1
        
        if p != -1:
            return arr[p]+h
        else:
            return h
```

```python
# https://leetcode.com/problems/word-break-ii/

# Given a string s and a dictionary of strings wordDict, add spaces in s to construct a sentence where each word is a valid dictionary word. Return all such possible sentences in any order.

# Note that the same word in the dictionary may be reused multiple times in the segmentation.

 

# Example 1:

# Input: s = "catsanddog", wordDict = ["cat","cats","and","sand","dog"]
# Output: ["cats and dog","cat sand dog"]

class Trie:
    def __init__(self):
        self.char_map = {}
    
    def insert(self, word):
        if len(word) == 0:
            self.char_map['<END>'] = Trie()
            
        else:
            if word[0] not in self.char_map:
                self.char_map[word[0]] = Trie()
            self.char_map[word[0]].insert(word[1:])
            
    def search(self, word):
        if len(word) == 0:
            return '<END>' in self.char_map
        
        if word[0] in self.char_map:
            return self.char_map[word[0]].search(word[1:])
        
        return False
    
class Solution(object):
    def dfs(self, s, trie, cache):
        if len(s) == 0:
            return []
        
        curr = ''
        out = []
        
        for i in range(len(s)):
            curr += s[i]
            p = trie.search(curr)
            if p:
                if i == len(s)-1:
                    out.append(curr)
                
                q = s[i+1:]
                x = self.dfs(q, trie, cache) if q not in cache else cache[q]
                    
                if len(x) > 0:
                    for y in x:
                        out.append(curr + ' ' + y)
        cache[s] = out
        return out
    
    def recursion(self, s, wordDict, i, cache):
        p = ''
        out = []
        
        for j in range(i, len(s)):
            p += s[j]
            if p in wordDict:
                a = self.recursion(s, wordDict, j+1, cache) if j+1 not in cache else cache[j+1]
                
                if len(a) > 0:
                    for x in a:
                        out.append(p + ' ' + x)
        
        cache[i] = out
        return out
                
                
    def wordBreak(self, s, wordDict):
        trie = Trie()
        
        for word in wordDict:
            trie.insert(word)
        
        return self.dfs(s, trie, {})
```

```python
# https://leetcode.com/problems/construct-binary-tree-from-preorder-and-inorder-traversal/

# Given two integer arrays preorder and inorder where preorder is the preorder traversal of a binary tree and inorder is the inorder traversal of the same tree, construct and return the binary tree.

class Solution(object):
    def build(self, inorder_pos_map, preorder, inorder, p, q, s, t):
        root = TreeNode(preorder[p])
        
        j = inorder_pos_map[preorder[p]]
        
        u = j-s
        
        if u > 0:
            root.left = self.build(inorder_pos_map, preorder, inorder, p+1, p+u, s, j-1)
        else:
            root.left = None
            
        if q-p-u > 0:
            root.right = self.build(inorder_pos_map, preorder, inorder, p+u+1, q, j+1, t)
        else:
            root.right = None
        
        return root
        
    def buildTree(self, preorder, inorder):
        inorder_pos_map = {}
        
        for i in range(len(inorder)):
            inorder_pos_map[inorder[i]] = i
        
        return self.build(inorder_pos_map, preorder, inorder, 0, len(preorder)-1, 0, len(inorder)-1)
        
```

```python
# https://leetcode.com/problems/clone-graph/

# Given a reference of a node in a connected undirected graph.

# Return a deep copy (clone) of the graph.

# Each node in the graph contains a value (int) and a list (List[Node]) of its neighbors.

# class Node {
#     public int val;
#     public List<Node> neighbors;
# }

class Solution(object):
    def cloneGraph(self, node):
        if node is None:
            return None
        
        queue = collections.deque([node])
        visited = set([node.val])
        adj_map, node_cache = {}, {}
        
        while len(queue) > 0:
            n = queue.popleft()
            adj_map[n.val] = []
            node_cache[n.val] = Node(n.val)
            
            for x in n.neighbors:
                adj_map[n.val].append(x.val)
                
                if x.val not in visited:
                    queue.append(x)
                    visited.add(x.val)
        
        for x, n in adj_map.items():
            for y in n:
                node_cache[x].neighbors += [node_cache[y]]
                
        return node_cache[node.val]
```

```python
# https://leetcode.com/problems/valid-palindrome-ii/

# Given a string s, return true if the s can be palindrome after deleting at most one character from it.

 

# Example 1:

# Input: s = "aba"
# Output: true

class Solution(object):
    def validPalindrome(self, s):
        i, j = 0, len(s)-1
        
        while i < j:
            if s[i] != s[j]:
                p = s[:i] + s[i+1:]
                q = s[:j] + s[j+1:]
                
                return p == p[::-1] or q == q[::-1]
            
            else:
                i += 1
                j -= 1
        
        return True
```

```python
# https://leetcode.com/problems/minimum-remove-to-make-valid-parentheses/

# Given a string s of '(' , ')' and lowercase English characters. 

# Your task is to remove the minimum number of parentheses ( '(' or ')', in any positions ) so that the resulting parentheses string is valid and return any valid string.

# Formally, a parentheses string is valid if and only if:

# It is the empty string, contains only lowercase characters, or
# It can be written as AB (A concatenated with B), where A and B are valid strings, or
# It can be written as (A), where A is a valid string.
 

# Example 1:

# Input: s = "lee(t(c)o)de)"
# Output: "lee(t(c)o)de"
# Explanation: "lee(t(co)de)" , "lee(t(c)ode)" would also be accepted.

class Solution(object):
    def minRemoveToMakeValid(self, s):
        stack = []
        for i in range(len(s)):
            if s[i] == ')':
                if len(stack) > 0 and s[stack[-1]] == '(':
                    stack.pop()
                else:
                    stack.append(i)
            elif s[i] == '(':
                stack.append(i)
        
        out = ''
        for i in range(len(s)-1, -1, -1):
            if len(stack) > 0 and stack[-1] == i:
                stack.pop()
            else:
                out = s[i] + out
        
        return out
```

```python
# https://leetcode.com/problems/minimum-moves-to-move-a-box-to-their-target-location/

# Storekeeper is a game in which the player pushes boxes around in a warehouse trying to get them to target locations.

# The game is represented by a grid of size m x n, where each element is a wall, floor, or a box.

# Your task is move the box 'B' to the target position 'T' under the following rules:

# Player is represented by character 'S' and can move up, down, left, right in the grid if it is a floor (empy cell).
# Floor is represented by character '.' that means free cell to walk.
# Wall is represented by character '#' that means obstacle  (impossible to walk there). 
# There is only one box 'B' and one target cell 'T' in the grid.
# The box can be moved to an adjacent free cell by standing next to the box and then moving in the direction of the box. This is a push.
# The player cannot walk through the box.
# Return the minimum number of pushes to move the box to the target. If there is no way to reach the target, return -1.

class Solution(object):
    def is_reachable(self, grid, u, v, p, q, m, n):
        heap = [(abs(u-p)+abs(v-q), u, v)]
        visited = set([(u, v)])
        
        while len(heap) > 0:
            x, i, j = heapq.heappop(heap)
            
            if (i, j) == (p, q):
                return True
            
            if i-1 >= 0 and grid[i-1][j] != '#' and (i-1, j) != (m, n) and (i-1, j) not in visited:
                heapq.heappush(heap, (abs(i-1-p)+abs(j-q), i-1, j)) 
                visited.add((i-1, j))
                
            if i+1 < len(grid) and grid[i+1][j] != '#' and (i+1, j) != (m, n) and (i+1, j) not in visited:
                heapq.heappush(heap, (abs(i+1-p)+abs(j-q), i+1, j)) 
                visited.add((i+1, j))
                
            if j-1 >= 0 and grid[i][j-1] != '#' and (i, j-1) != (m, n) and (i, j-1) not in visited:
                heapq.heappush(heap, (abs(i-p)+abs(j-1-q), i, j-1)) 
                visited.add((i, j-1))
                
            if j+1 < len(grid[0]) and grid[i][j+1] != '#' and (i, j+1) != (m, n) and (i, j+1) not in visited:
                heapq.heappush(heap, (abs(i-p)+abs(j+1-q), i, j+1)) 
                visited.add((i, j+1))
                
        return False
            
    def minPushBox(self, grid):
        b_x, b_y = 0, 0
        s_x, s_y = 0, 0
        t_x, t_y = 0, 0
        
        for i in range(len(grid)):
            for j in range(len(grid[0])):
                if grid[i][j] == 'B':
                    b_x, b_y = i, j
                    
                if grid[i][j] == 'S':
                    s_x, s_y = i, j
                    
                if grid[i][j] == 'T':
                    t_x, t_y = i, j
                    
                    
        heap = [(abs(b_x-t_x)+abs(b_y-t_y), 0, b_x, b_y, s_x, s_y)]
        cache = {}
        min_steps = float("Inf")
        
        while len(heap) > 0:
            x, steps, p, q, u, v = heapq.heappop(heap)
            
            if (p, q) == (t_x, t_y):
                min_steps = min(min_steps, steps)
                
            elif steps < min_steps:
                if p-1 >= 0 and grid[p-1][q] != '#' and self.is_reachable(grid, u, v, p+1, q, p, q):
                    if (p-1, q, p, q) not in cache or cache[(p-1, q, p, q)] > steps+1: 
                        heapq.heappush(heap, (abs(p-1-t_x)+abs(q-t_y)+steps+1, steps+1, p-1, q, p, q))
                        cache[(p-1, q, p, q)] = steps+1
                
                if p+1 < len(grid) and grid[p+1][q] != '#' and self.is_reachable(grid, u, v, p-1, q, p, q):
                    if (p+1, q, p, q) not in cache or cache[(p+1, q, p, q)] > steps+1: 
                        heapq.heappush(heap, (abs(p+1-t_x)+abs(q-t_y)+steps+1, steps+1, p+1, q, p, q))
                        cache[(p+1, q, p, q)] = steps+1
                        
                if q-1 >= 0 and grid[p][q-1] != '#' and self.is_reachable(grid, u, v, p, q+1, p, q):
                    if (p, q-1, p, q) not in cache or cache[(p, q-1, p, q)] > steps+1:
                        heapq.heappush(heap, (abs(p-t_x)+abs(q-1-t_y)+steps+1, steps+1, p, q-1, p, q))
                        cache[(p, q-1, p, q)] = steps+1
                        
                if q+1 < len(grid[0]) and grid[p][q+1] != '#' and self.is_reachable(grid, u, v, p, q-1, p, q):
                    if (p, q+1, p, q) not in cache or cache[(p, q+1, p, q)] > steps+1: 
                        heapq.heappush(heap, (abs(p-t_x)+abs(q+1-t_y)+steps+1, steps+1, p, q+1, p, q))
                        cache[(p, q+1, p, q)] = steps+1
                        
        return min_steps if min_steps != float("Inf") else -1
```

```python
# https://leetcode.com/problems/verifying-an-alien-dictionary/

# In an alien language, surprisingly they also use english lowercase letters, but possibly in a different order. The order of the alphabet is some permutation of lowercase letters.

# Given a sequence of words written in the alien language, and the order of the alphabet, return true if and only if the given words are sorted lexicographicaly in this alien language.

 

# Example 1:

# Input: words = ["hello","leetcode"], order = "hlabcdefgijkmnopqrstuvwxyz"
# Output: true
# Explanation: As 'h' comes before 'l' in this language, then the sequence is sorted.

class Solution(object):
    def is_smaller(self, a, b, pos_map):
        i = 0
        while i < min(len(a), len(b)) and a[i] == b[i]:
            i += 1
        
        if i == len(a):
            return True
        elif i == len(b):
            return False
        else:
            return pos_map[a[i]] < pos_map[b[i]]
        
    def isAlienSorted(self, words, order):
        if len(words) == 1:
            return True
        
        pos_map = {}
        for i in range(len(order)):
            pos_map[order[i]] = i
        
        for i in range(len(words)-1):
            f = self.is_smaller(words[i], words[i+1], pos_map)
            if f is False:
                return False
        
        return True
```

```python
# https://leetcode.com/problems/lowest-common-ancestor-of-a-binary-tree/

# Given a binary tree, find the lowest common ancestor (LCA) of two given nodes in the tree.

# According to the definition of LCA on Wikipedia: “The lowest common ancestor is defined between two nodes p and q as the lowest node in T that has both p and q as descendants (where we allow a node to be a descendant of itself).”

class Solution(object):
    def traverse(self, root, p, q):
        if root is None:
            return False, False, None
        
        a, u, x = self.traverse(root.left, p, q)
        b, v, y = self.traverse(root.right, p, q)
        
        if a:
            return a, u, x
        if b:
            return b, v, y
        
        if u and v:
            return True, True, root
        
        if root.val == p.val or root.val == q.val:
            if u or v:
                return True, True, root
            return False, True, None
        
        return False, u or v, None
    
    def lowestCommonAncestor(self, root, p, q):
        return self.traverse(root, p, q)[2]
```

```python
# https://leetcode.com/problems/split-array-largest-sum/description/

# Given an array nums which consists of non-negative integers and an integer m, you can split the array into m non-empty continuous subarrays.

# Write an algorithm to minimize the largest sum among these m subarrays.

 

# Example 1:

# Input: nums = [7,2,5,10,8], m = 2
# Output: 18
# Explanation:
# There are four ways to split nums into two subarrays.
# The best way is to split it into [7,2,5] and [10,8],
# where the largest sum among the two subarrays is only 18.

class Solution(object):
    def num_parts(self, nums, k):
        queue = collections.deque([])
        queue_sum = 0
        max_queue_sum = -float("Inf")
        c = 0
        
        for i in range(len(nums)):
            queue.append(nums[i])
            queue_sum += nums[i]
            
            if queue_sum <= k:
                max_queue_sum = max(max_queue_sum, queue_sum)
            else:
                c += 1
                queue = collections.deque([nums[i]])
                queue_sum = nums[i]
        
        if len(queue) > 0:
            c += 1
            max_queue_sum = max(max_queue_sum, queue_sum)
        
        return c, max_queue_sum
                
                
    def splitArray(self, nums, m):
        left, right = max(nums), sum(nums)
        p = -1
        
        while left <= right:
            mid = (left + right)/2
            
            q, x = self.num_parts(nums, mid)
            
            if q <= m:
                p = x
                right = mid-1
            else:
                left = mid+1
                
        return p
```

```python
# https://leetcode.com/problems/path-with-minimum-effort/

# You are a hiker preparing for an upcoming hike. You are given heights, a 2D array of size rows x columns, where heights[row][col] represents the height of cell (row, col). You are situated in the top-left cell, (0, 0), and you hope to travel to the bottom-right cell, (rows-1, columns-1) (i.e., 0-indexed). You can move up, down, left, or right, and you wish to find a route that requires the minimum effort.

# A route's effort is the maximum absolute difference in heights between two consecutive cells of the route.

# Return the minimum effort required to travel from the top-left cell to the bottom-right cell.

class Solution(object):
    def minimumEffortPath(self, heights):
        heap = [(0, 0, 0)]
        cache = {(0, 0): 0}
        
        min_cost = float("Inf")
        
        while len(heap) > 0:
            cost, i, j = heapq.heappop(heap)
            
            if (i, j) == (len(heights)-1, len(heights[0])-1):
                min_cost = min(min_cost, cost)
            
            elif cost < min_cost:
                p = [[-1,0],[0,-1],[1,0],[0,1]]

                for a, b in p:
                    if 0 <= i+a < len(heights) and 0 <= j+b < len(heights[0]):
                        new_cost = max(cost, abs(heights[i+a][j+b]-heights[i][j]))

                        if (i+a, j+b) not in cache or cache[(i+a, j+b)] > new_cost:
                            heapq.heappush(heap, (new_cost, i+a, j+b))
                            cache[(i+a, j+b)] = new_cost
                            
        return min_cost
```

```python
# https://leetcode.com/discuss/interview-question/124827/Facebook-or-Minimum-number-of-people-to-spread-a-message

# Considering that I'ld would like to spread a promotion message across all people in twitter. Assuming the ideal case, if a person tweets a message, then every follower will re-tweet the message.

# You need to find the minimum number of people to reach out (for example, who doesn't follow anyone etc) so that your promotion message is spread out across entire network in twitter.

# Also, we need to consider loops like, if A follows B, B follows C, C follows D, D follows A (A -> B -> C -> D -> A) then reaching only one of them is sufficient to spread your message.

# Input: A 2x2 matrix like below. In this case, a follows b, b follows c, c follows a.

#     a b c
# a  1 1 0
# b  0 1 1
# c  1 0 1
# Output: List of people to be reached to spread out message across everyone in the network.
                                                            
def dfs(adj_map, i, visited, senders):
    for j in adj_map[i]:
        if j not in visited:
            visited.add(j)
            if j in senders:
                senders.remove(j)
            dfs(adj_map, j, visited, senders)
                
def minimum_spread_message(n, edges):
    adj_map = [[] for i in range(n)]
    
    for x, y in edges:
        adj_map[x].append(y)
    
    senders = set()
    all_visited = set()
    
    min_senders = n
    
    for i in range(n):
        senders.add(i)
        visited = set([i])
        dfs(adj_map, i, visited, senders)
        all_visited.update(visited)
        
        if len(all_visited) == n:
            min_senders = min(min_senders, len(senders))
    
    return min_senders

edges = [[3,0],[3,4],[5,0],[5,1],[5,2],[5,4],[0,1],[2,3],[4,5]]
minimum_spread_message(6, edges)
```

```python
# https://leetcode.com/problems/sort-colors/

# Given an array nums with n objects colored red, white, or blue, sort them in-place so that objects of the same color are adjacent, with the colors in the order red, white, and blue.

# We will use the integers 0, 1, and 2 to represent the color red, white, and blue, respectively.

# You must solve this problem without using the library's sort function.

 

# Example 1:

# Input: nums = [2,0,2,1,1,0]
# Output: [0,0,1,1,2,2]

class Solution(object):
    def sortColors(self, nums):
        one_first_index, two_first_index = float("Inf"), float("Inf")
        
        i = 0
        while i < len(nums):
            if nums[i] == 2:
                two_first_index = min(two_first_index, i)
                
            elif nums[i] == 1:
                if two_first_index < i:
                    nums[two_first_index] = 1
                    nums[i] = 2
                    one_first_index = min(one_first_index, two_first_index)
                    two_first_index += 1
                    
                one_first_index = min(one_first_index, i)
            
            else:
                if one_first_index < i and two_first_index < i:
                    nums[one_first_index] = 0
                    nums[two_first_index] = 1
                    nums[i] = 2
                    
                    one_first_index += 1
                    two_first_index += 1
                    
                elif one_first_index < i:
                    nums[one_first_index] = 0
                    nums[i] = 1
                    one_first_index += 1
                    
                elif two_first_index < i:
                    nums[two_first_index] = 0
                    nums[i] = 2
                    two_first_index += 1
            
            i += 1
```

```python
# https://www.lintcode.com/problem/663/

# You are given a m x n 2D grid initialized with these three possible values.

# -1 - A wall or an obstacle.
# 0 - A gate.
# INF - Infinity means an empty room. We use the value 2^31 - 1 = 2147483647 to represent INF as you may assume that the distance to a gate is less than 2147483647.
# Fill each empty room with the distance to its nearest gate. If it is impossible to reach a Gate, that room should remain filled with INF

class Solution:
    """
    @param rooms: m x n 2D grid
    @return: nothing
    """
    def wallsAndGates(self, rooms):
        queue = collections.deque([])
        visited = set()

        for i in range(len(rooms)):
            for j in range(len(rooms[0])):
                if rooms[i][j] == 0:
                    queue.append((i, j, 0))
                    visited.add((i, j))

        while len(queue) > 0:
            i, j, d = queue.popleft()

            if rooms[i][j] == 2147483647:
                rooms[i][j] = d
            
            p = [[-1,0],[0,-1],[1,0],[0,1]]

            for x, y in p:
                if 0 <= i+x < len(rooms) and 0 <= j+y < len(rooms[0]) and (i+x, j+y) not in visited and rooms[i+x][j+y] != -1:
                    queue.append((i+x, j+y, d+1))
                    visited.add((i+x, j+y))
        
        return rooms
```

```python
# https://www.lintcode.com/problem/1534/

# Convert a BST to a sorted circular doubly-linked list in-place. Think of the left and right pointers as synonymous to the previous and next pointers in a doubly-linked list.

"""
Definition of TreeNode:
class TreeNode:
    def __init__(self, val):
        self.val = val
        self.left, self.right = None, None
"""
class Solution:
    """
    @param root: root of a tree
    @return: head node of a doubly linked list
    """
    def bst_to_dll(self, root):
        if root is None:
            return None, None

        a_h, a_t = self.bst_to_dll(root.left)
        b_h, b_t = self.bst_to_dll(root.right)

        node = ListNode(root.val)
        h, t = a_h, b_t

        if a_t is not None:
            a_t.next = node
        else:
            h = node

        node.prev = a_t

        node.next = b_h

        if b_h is not None:
            b_h.prev = h
        else:
            t = node

        return h, t

    def treeToDoublyList(self, root):
        if root is None:
            return None

        h, t = self.bst_to_dll(root)
        t.next = h
        h.prev = t
        return h
```

```python
# https://leetcode.com/discuss/interview-question/124832/Facebook-or-Design-data-structure-for-wildcard-search

# Given such data structure, implement add() and search() method, which support wildcard matching.
# ? Matches any single character.
# * Matches any sequence of characters (including the empty sequence).

# class Node 
# {
# Node getChildForLetter(char letter)
# Node[] getAllChildren();
# boolean isTerminal();
# }

# //some example:
# add("car")
# add("caw")
# add("cauw")

# search(“c*w”) 
# //should return “caw” and “cauw”.
# // "*" could be at any place in the input string.


class Trie:
    def __init__(self, val):
        self.val = val
        self.char_map = {}
    
    def insert(self, word, orig):
        if len(word) == 0:
            self.char_map['$'] = Trie(orig)
        else:
            if word[0] not in self.char_map:
                self.char_map[word[0]] = Trie('')
            
            self.char_map[word[0]].insert(word[1:], orig)
    
    def search(self, word, i):
        out = []
        
        if i == len(word):
            if '$' in self.char_map:
                return [self.char_map['$'].val]
            return []
        
        if word[i] == '?':
            for x in self.char_map:
                if x != '$':
                    out += self.char_map[x].search(word, i+1)
                
        elif word[i] == '*':
            for x in self.char_map:
                if x != '$':
                    out += self.char_map[x].search(word, i)
                    out += self.char_map[x].search(word, i+1)
        
        else:
            if word[i] in self.char_map:
                out += self.char_map[word[i]].search(word, i+1)
        
        return out
    
trie = Trie('')
trie.insert('cat', 'cat')
trie.insert('cart', 'cart')
trie.insert('cartography', 'cartography')
trie.insert('capolttt', 'capolttt')

trie.search('c*t', 0)
```

```python
# https://leetcode.com/discuss/interview-question/124656/Uber-phone-screen-String-cache

# FIFO Cache of N Strings. Both methods should run in constant time.

# class Cache {
#     final int N;
#     void add (String element) {
#     // Implement constant-time here.
#     {
#     boolean contains (String element) {
#     // Implement constant-time here.
#     }
# }
# Curve-ball to complicate the implementation. Implement remove in constant time.

# void remove (String element) {
# // Implement constant-time here.
# }
# Do not ignore duplicate entries.
# If cache size N is reached, FIFO implementation should pop the first element from the cache for every new entry.
        
        
class fifocache:
    def __init__(self, n):
        self.prev = {}
        self.next = {}
        self.seq_map = {}
        self.head = None
        self.tail = None
        self.max_size = n
        self.cur_size = 0
        self.seq = 0
    
    def add(self, x):
        if self.head is None:
            self.prev[(x, self.seq)] = None
            self.next[(x, self.seq)] = None
            
            self.head = (x, self.seq)
            self.tail = (x, self.seq)
            
        else:
            if self.cur_size == self.max_size:
                self.remove(self.tail[0])
            
            self.prev[(x, self.seq)] = None
            self.next[(x, self.seq)] = self.head
            self.prev[self.head] = (x, self.seq)
            self.head = (x, self.seq)
            
        if x not in self.seq_map:
            self.seq_map[x] = []
        self.seq_map[x].append(self.seq)
        
        self.cur_size += 1
        self.seq += 1
    
    def contains(self, x):
        return x in self.seq_map
    
    def remove(self, x):
        if x in self.seq_map:
            u = self.seq_map[x].pop()
            p = (x, u)
            
            if self.next[p] is None:
                self.tail = self.prev[p]
            else:
                self.prev[self.next[p]] = self.prev[p]
                
            if self.prev[p] is None:
                self.head = self.next[p]
            else:
                self.next[self.prev[p]] = self.next[p]
            
            self.prev.pop(p)
            self.next.pop(p)
            
            if len(self.seq_map[x]) == 0:
                self.seq_map.pop(x)
            
            self.cur_size -= 1
```
